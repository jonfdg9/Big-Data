#DISTRIBUCION NORMAL CON PYTHON



La distribución normal (o distribución de Gauss) es una de las distribuciones más importantes en estadística, caracterizada por una curva en forma de campana simétrica, donde la media, la mediana y la moda coinciden. Los datos tienden a agruparse en torno a la media, y la probabilidad disminuye conforme nos alejamos de ella. Se describe por dos parámetros:

**Media** (μ): Ubica el centro de la distribución.

**Desviación estándar** (σ): Mide la dispersión de los datos alrededor de la media.

DistribucionNormal.png

##Probabilidades


P(X > x1): Esta probabilidad corresponde al área bajo la curva a la derecha del valor x1. Es la probabilidad de que la variable aleatoria X tome un valor mayor a x1.

P(X ≤ x1): Es la probabilidad acumulada hasta X, también conocida como la función de distribución acumulativa (CDF, por sus siglas en inglés). Representa el área bajo la curva a la izquierda de x1.

P(x1 < X < x2): Esta probabilidad corresponde al área entre x1 y x2 bajo la curva. Es la probabilidad de que X tome un valor entre x1 y x2.



##Relación entre probabilidades

P(X > x1) = 1 − P(X ≤ x1): La probabilidad de que X sea mayor que x1 es simplemente 1 menos la probabilidad de que X sea menor o igual que x1, ya que estas dos probabilidades cubren el total del área bajo la curva (que suma 1).

P(x1 < X < x2) = P(X ≤ x2)−P (X ≤ x1): Para calcular la probabilidad de que X esté entre x1 y x2, restamos a la  de la probabilidad acumulada en x2 la probabilidad acumulada en x1.


## Ejemplo1 con Python

Se sabe que la cantidad de horas de sueño de una persona sigue una distribución normal con una media de 7 horas y una desviación estándar de 1.5 horas. Con base en esta información, responde lo siguiente:

1. ¿Cuál es la probabilidad de que una persona duerma menos de 3 horas?
2. ¿Cuál es la probabilidad de que una persona duerma más de 10 horas seguidas?
3. ¿Cuál es la probabilidad de que una persona duerma entre 5 y 7 horas?





Importamos la librería necesaria para poder trabajar con la distribución normal

```python
import scipy.stats as stats

```

Definimos la media y la desviación estándard de la distribución normal

```python
# Parámetros de la distribución normal
mu = 7    # Media
sigma = 1.5  # Desviación estándar
```

Calculamos la probabilidad de que una persona al azar duerma menos de 3 horas

```python
# 2. Probabilidad de dormir menos de 3 horas
prob_menos_3 = stats.norm.cdf(3, mu, sigma)
print(f"Probabilidad de dormir menos de 3 horas es: {prob_menos_3:.4f}")
```

Calculamos la probabilidad de que una persona al azar duerma más de 10 horas seguidas

```python
# 1. Probabilidad de dormir más de 10 horas
prob_mas_10 = 1 - stats.norm.cdf(10, mu, sigma)
print(f"Probabilidad de dormir más de 10 es: {prob_mas_10:.4f}")
```

Calculamos la probabilidad de que una persona al azar duerma entre 5 y 7 horas

```python
# 3. Probabilidad de dormir entre 5 y 7 horas
prob_entre_5_7 = stats.norm.cdf(7, mu, sigma) - stats.norm.cdf(5, mu, sigma)

# Mostrar resultados
print(f"Probabiliad de dormir entre 5 y 7 horas es: {prob_entre_5_7:.4f}")
```

## Ejemplo 2 con Python

Supongamos que la altura de las margaritas en un jardín sigue una distribución normal con una media de 15 centímetros y una desviación estándar de 2 centímetros.

1. ¿Cuál es la probabilidad de que una margarita elegida al azar mida exactamente 18 centímetros?

2.  ¿Cuál es la altura mínima que debe tener una margarita para que el 90% de las margaritas sean más bajas que ella?

Importamos las librerías necesarias


```python
import numpy as np
from scipy import stats
```

 Definimos la media y la desviación estándar de la función de densidad de probabilidad

```python

media = 15  # Media de la altura en centímetros
desviacion_estandar = 2  # Desviación estándar en centímetros
```

1. ¿Cuál es la probabilidad de que una margarita elegida al azar mida exactamente 18 centímetros?

```python
# Valor para el que queremos calcular la PDF
altura = 18  # Altura en centímetros

# Calculando la PDF
probabilidad = stats.norm.pdf(altura, loc=media, scale=desviacion_estandar)

print(f"La probabilidad de que una margarita mida exactamente 18 cm es: {probabilidad:.4f}")
```

 2.- ¿Cuál es la altura mínima que debe tener una margarita para que el 90% de las margaritas sean más bajas que ella?

 Estamos buscando un valor de altura tal que el 90% de las margaritas estén por debajo de esa altura. En términos estadísticos, estamos buscando el percentil 90 de la distribución.

```python
# Percentil que queremos encontrar (90%)
percentil = 0.9

# Calculando la altura correspondiente al percentil
altura_minima = stats.norm.ppf(percentil, loc=media, scale=desviacion_estandar)

print(f"La altura mínima para que el 90% de las margaritas sean más bajas es: {altura_minima:.4f} cm.")
```
