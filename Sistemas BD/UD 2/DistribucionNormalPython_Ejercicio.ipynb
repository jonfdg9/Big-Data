```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

mu = 1.9   # Media
sigma = 0.1 # Desviaci칩n est치ndar

prob= stats.norm.cdf(1.85, mu, sigma)
prob2=1- stats.norm.cdf(2, mu, sigma)
print(f"{1-(prob+prob2)}")



x=np.linspace(mu - 3* sigma, mu +3* sigma)

f1=stats.norm.pdf(x,loc=mu,scale=sigma)

plt.figure(figsize=(8, 6))
plt.plot(x, f1)

plt.show()
```
